```
from zipfile import ZipFile
import pandas as pd
```

```
with ZipFile('usa-information.zip', 'r') as data:
   # Get a list of all archived file names from the zip
   listNames = data.namelist()
   # Iterate over the file names
   for fileName in listNames:
       # Check filename endswith csv
       if fileName.endswith('.csv'):
           # Extract a single file from zip
           data.extract(fileName, 'temp_csv')
```

```
lifeexpect = pd.read_csv('Lifeexpentacy.csv')
lifeexpect = lifeexpect.rename(columns={'State/federal district/territory':'States'})
# drop columns
lifeexpect = lifeexpect.drop(['Rank','Life Expectancy, White (in years)',
                              'Life Expectancy, African American (in years)',
                              'Life Expectancy, Asian American (in years)',
                              'Life Expectancy, Latino (in years)','Life Expectancy, Native American (in years)'],
                              axis = 1)
lifeexpect = lifeexpect.rename(columns={'Life Expectancy, All(in years)':'Life Expectancy (years)'})
# drop rows
lifeexpect = lifeexpect.drop([2,23,27,45,48,53,56]) 
# sort the dataframe by the column values
lifeexpect = lifeexpect.sort_values(by='States')
#reset the index of the df according to the sorted values 
lifeexpect.reset_index(drop=True, inplace=True) 
print(lifeexpect.shape)
lifeexpect.head(3)
```

```
population = pd.read_csv('Population.csv')
# drop columns
population = population.drop(['Rank in the fifty states, 2018', 'Rank in states & territories',
                              'Census population, April 1, 2010[6]', 'Percent change from 2010-2018[note 1]',
                              'Total seats in the U.S. House of Representatives, 2013â€“2023',
                              'Estimated population per electoral vote, 2018[note 2]','Estimated population per House seat, 2018',
                              'Census population per House seat, 2010',
                              'Percent of the total U.S. population, 2018[note 3]',], axis = 1)
# rename columns
population = population.rename(columns={'Name':'State','Population estimate, July 1, 2018[5]':'Population estimate 2018'})
population = population.drop([2,9,12,37,42,48,56,57,58,59]) #drop rows
population = population.sort_values(by='State') # sort values
population.reset_index(drop=True, inplace=True) #reset index
print(population.shape)
population.head(3)
```

```
education = pd.read_csv('education.csv')
# drop columns
education = education.drop(['Rank', 'Rank.1', 'Rank.2'], axis = 1)
education = education.drop([16,34]) # drop rows
education = education.sort_values(by='State')# sort values
education.reset_index(drop=True, inplace=True) # reset index
print(education.shape)
education.head(3)
```

```
# Create a new dataframe by combining the sorted dataframes
df = pd.concat([lifeexpect, education, population], axis=1)
df = df.drop(['State', 'State'], axis=1)
print(df.shape)
df.head()
```

```
# Clean the values in the dataframe
# strips the '%' from each value in the column and turns the value into float
def str_clean(cell_contents):
  return float(cell_contents.strip('%')) 
  
new_list = []
for cell in df['% High school graduate']:
  new_list.append(str_clean(cell))
df['% High school graduate'] = pd.Series(new_list)

new_list = []
for cell in df["% Bachelor's degree"]:
  new_list.append(str_clean(cell))
df["% Bachelor's degree"] = pd.Series(new_list)

new_list = []
for cell in df["% Advanced degree"]:
  new_list.append(str_clean(cell))
df["% Advanced degree"] = pd.Series(new_list)
```

```
# strips the '[]' and ',' from each value in the column and turns the value into float
def str_clean(cell_contents):
  return float(cell_contents.replace(',','').replace('[','').replace(']',''))

new_list = []
for cell in df["Population estimate 2018"]:
  new_list.append(str_clean(cell))
df["Population estimate 2018"] = pd.Series(new_list)
```

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(15, 12))

labels = list(df['States'])
data2 = list(df["% Bachelor's degree"])
data3 = list(df['% Advanced degree'])

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars


rects1 = ax.bar(x - width/2, data2, width, label="Bachelor's degree", color = '#BDC3D5')
rects2 = ax.bar(x + width/2, data3, width, label='Advanced degree', color = '#8856A7')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Percent of Population', color="black", fontsize=14)
ax.set_title('Higher Education by State', color="black", weight = 'bold', fontsize=17)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation = 'vertical')
ax.legend()

ax.grid(None) #remove the grid
ax.set_facecolor('white') #make the background color 'white'

# attach a text label above each bar in *rects*, displaying its height
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
fig.set_facecolor('#E5E5E5')

plt.show()
```

```
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt

xs = df["% Bachelor's degree"]
ys = df['Life Expectancy (years)']

def best_fit(xs, ys):  # function to calculate slope and y-intercept
  m = (((mean(xs) * mean(ys)) - mean(xs * ys)) / 
      ((mean(xs) * mean(xs)) - mean(xs * xs)))
  b = mean(ys) - m * mean(xs)
  return m, b

m, b = best_fit(xs, ys)  # assign m & b to function's output
print(m,b)
```

```
df['regression_line'] = [(m * x) + b for x in xs] # calculate the regression line
```

```
ST=pd.DataFrame({"State" : ["AL","AK","AZ","AR","CA","CO","CT","DE","FL","GA",
        "HI","ID","IL","IN","IA","KS","KY","LA","ME","MD",
        "MA","MI","MN","MS","MO","MT","NE","NV","NH","NJ","NM","NY","NC",
        "ND","OH","OK","OR","PA","RI","SC","SD","TN","TX","UT","VT","VA","WA","WV","WI","WY"]})
RG = df['regression_line']
new_df = pd.concat([ST,RG], axis =1) #create the data for later use in folium map
new_df.head(3)
```

```
import folium
from urllib.request import urlopen
import json # import the map of the states
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
state_geo = f'{url}/us-states.json'

# 'location' centers the map on the area you need
m = folium.Map(location=[40, -97], zoom_start=4, width=700, height=500, control_scale = False)

folium.Choropleth(
    geo_data=state_geo,
    name='choropleth',
    data=new_df,
    columns=['State','regression_line'],
    key_on='id',
    fill_color='BuPu',
    fill_opacity=0.6,
    line_opacity=0.6,
    legend_name='Life Expectancy',).add_to(m)

folium.LayerControl().add_to(m)

m.save(outfile='map.html')
m
```
